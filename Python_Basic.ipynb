1. What is Python, and why is it popular?

    -> Python is a high-level, interpreted programming language that is easy to read and write. It uses simple English-like syntax, so beginners can learn it quickly

    It is popular because:

    1) It is easy to learn and understand.

    2) It has a large number of libraries for AI, data science, and web development.

2. What is an interpreter in Python?

    -> An interpreter in Python is a program that reads and runs the Python code line by line.
    It converts each line of code into machine language so the computer can understand it.


3.  What are pre-defined keywords in Python?

    ->	Pre-defined keywords in Python are special reserved words that have a specific meaning and purpose in the language.

4. Can keywords be used as variable names?

   ->	No, keywords cannot be used as variable names in Python.

5.  What is mutability in Python?

    ->	Mutability means whether a data object can be changed after it is created.

    -If an object is mutable, we can change its contents without creating a new object.

    -If it is immutable, its contents cannot be changed once it is created.

6.  Why are lists mutable, but tuples are immutable?

    -> 	Lists are mutable because we can change, add, or remove their elements after creating them.
    Python stores lists in a way that allows modification of their contents.

    -Tuples are immutable because their elements cannot be changed once they are created. Python fixes the memory for a tuple, so it becomes faster and more secure for data that should not change.

7. What is the difference between “==” and “is” operators in Python?

    -> “==” checks if the values of two objects are the same.

    “is” checks if both variables refer to the same object in memory.

8.  What are logical operators in Python?

    -> Logical operators in Python are used to combine conditional statements.

    They return either True or False based on the conditions.

    There are three logical operators:

    **and** → True only if both conditions are True

    **or**→ True if at least one condition is True

    **not** → reverses the result (True becomes False, and vice versa)

9.  What is type casting in Python?

     ->	Type casting in Python means converting one data type into another.

      For example, we can change a string into an integer or a float.

      Python provides built-in functions for this, like **int()**, **float()**, **str()**, and **bool()**.

10.  What is the difference between implicit and explicit type casting?

     ->	Implicit type casting happens automatically by Python.
     Python converts one data type to another without the programmer writing code for it.

     -Explicit type casting is done manually using functions like **int(), float(), or str().**

11.  What is the purpose of conditional statements in Python?

     -> Conditional statements are used to make decisions in a program.

     They let the program choose different actions based on whether a condition is True or False.

12.  How does the elif statement work?

     ->	The elif statement in Python means "else if".

     It allows you to check multiple conditions, one after another.
     If one condition is True, Python runs that block and skips the rest.

13.  What is the difference between for and while loops in Python?

     ->	A for loop runs a fixed number of times.
      It is used when we know how many times we want to repeat something.
      A while loop runs as long as a condition is True.
      It is used when we don’t know exactly how many times the loop should run.

14.  Describe a scenario where a while loop is more suitable than a for loop.

     -> A while loop is better when we don’t know how many times the loop needs to run.
     It keeps running until a certain condition becomes False.

```python
# 09) Write a Python program to convert user input from string to integer, float, and boolean types.
user_input = input("Enter any number: ")

# Convert string to integer
int_value = int(user_input)
print("Integer value:", int_value)

# Convert string to float
float_value = float(user_input)
print("Float value:", float_value)

# Convert string to boolean
bool_value = bool(user_input)
print("Boolean value:", bool_value)
```

```python
# 10) Write code to demonstrate type casting with list elements.
string_list = ["10", "20", "30", "40"]

print("Original list:", string_list)

# Convert each element to integer
int_list = [int(x) for x in string_list]
print("After converting to integers:", int_list)

# Convert each element to float
float_list = [float(x) for x in string_list]
print("After converting to floats:", float_list)
```

```python
# 13) Write a Python program to find the sum of all even numbers between 1 and 50.
sum_even = 0

for i in range(1, 51):
    if i % 2 == 0:
        sum_even += i

print("The sum of all even numbers between 1 and 50 is:", sum_even)
```

```python
# 14) Write a program to reverse a string using a while loop.
text = input("Enter a string: ")
reversed_text = ""
index = len(text) -1
while index >=0:
  reversed_text += text[index]
  index -= 1


print("Reversed String:", reversed_text)
```

```python
# 15) Write a Python program to calculate the factorial of a number provided by the user using a while loop
num = int(input("Enter the number: "))
factorial = 1
i = 1
while i<= num:
  factorial *= i
  i += 1

print("The factorial of",num,"is:",factorial)
```
